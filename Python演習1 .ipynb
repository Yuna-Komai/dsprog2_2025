### 問1

文字列"datascience"の文字を逆に（末尾から先頭に向かって）並べた文字列を出力してください

```python
s = "datascience"
# reverse で文字列を逆順にする
reversed_s = s[::-1]
print(reversed_s)
```

### 問2

「ヘエルスシトンニキア」という文字列の1,3,5,7,9文字目を取り出して連結した文字列を出力してください


```python
s = "ヘエルシトシニキア"
# 2文字おきに抽出する
# スライスは [開始位置:終了位置:ステップ] 
# 開始位置と終了位置を省略すると、文字列全体が対象になる
result = s[::2]
print(result)
```

### 問3

「ヘルシンキ」＋「エストニア」の文字を先頭から交互に連結して文字列「ヘエルスシトンニキア」を出力してください


```python
s1 = "ヘルシンキ"
s2 = "エストニア"

# zipで2つの文字列をペアにし、各ペアから文字を取り出してリストにする
# その後、joinで結合して1つの文字列にする
result = "".join([c1 + c2 for c1, c2 in zip(s1, s2)])
print(result)
```

### 問4

"Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成してください


```python
sentence = "Character education based on the fundamental spirit of Buddhism and the Shiguizeigan"
# 単語に分解する
words = sentence.split()
# 前から順に各単語の文字数（長さ）を取得する
word_lengths = [len(word) for word in words]
print(word_lengths)
```

### 問5

"It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."という文を単語に分解し，1, 2, 5, 6, 8, 9, 16番目の単語の先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）を格納した辞書を作成してください

```python
sentence = "It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."
# 句読点を除去してから単語に分解する
clean_sentence = sentence.replace(",", "").replace(".", "")
words = clean_sentence.split()

# 取り出す単語の位置
positions = [1, 2, 5, 6, 8, 9, 16]

# 結果を格納する場所
result_dict = {}

# 指定された位置の単語を処理
for pos in positions:
    # リストのインデックスは0から始まるから1を引く
    index = pos - 1
    
    # 単語の先頭2文字を取り出す
    two_chars = words[index][:2]
    
    # 格納
    result_dict[two_chars] = pos

print(result_dict)
```
